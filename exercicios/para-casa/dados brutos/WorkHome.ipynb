Explorando a questão de consumo e venda pelas cidades do Brasil, colocando um top 5 de cada pesquisa feita.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Usando o geolocation
# eu ACHO que pesquisei de todas as localizações de quais cidades mais tem usuários conectados a base
df_geolocation['geolocation_city'].value_counts(normalize=True).head(10)
#São Paulo o maior com duas posições sendo a primeira com 0.135778 e a segunda aparição em 4º lugar com 0.024914 
```

```python
plt.figure(figsize=(8, 6))

# instanciar o tipo de plotagem, no caso, um countplot usando x como as faixas etárias
category = ['SP', 'Curitiba', 'RJ', 'BH', 'Ribeirão P', 'Guarulhos', 'Ibitinga', 'Santo Andre', 'Campinas', 'Maringá']
values = [0.224233, 0.041034, 0.031018, 0.021971, 0.016801, 0.016155, 0.015832, 0.014540, 0.013247, 0.012924]

plt.bar(category, values, color ='#fd7d00')
# definir texto do título e eixos x e y
plt.title('User or computing device via a variety of data collection mechanisms')
plt.xlabel('City')
plt.ylabel('Scale of geolocaion')

plt.show()
```

```python
#Estdo com maior quantidade de vendedores
df_sellers['seller_city'].value_counts(normalize=True).head(10)
#São Paulo com 0.224233
```
