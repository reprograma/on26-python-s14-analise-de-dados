```python
import pandas as pd 
```

COLETAS DE DADOS

```python
df_avaliacao = pd.read_csv('datasets_olist/olist_order_reviews_dataset.csv')
df_avaliacao
```

```python
df_ordem = pd.read_csv('datasets_olist/olist_orders_dataset.csv')
df_ordem
```

```python
df_ordem_items = pd.read_csv('datasets_olist/olist_order_items_dataset.csv')
df_ordem_items
```

JUNTANDO OS DATASETS

```python
df_olist = pd.merge(df_avaliacao, df_ordem, on='order_id')
df_olist
```

```python
df_olist = pd.merge(df_olist, df_ordem_items, on='order_id')
df_olist
```

```python
#minha base final criada

df_olist.to_csv('base_final_s14_olist_marcinha.csv', index=False)
```
