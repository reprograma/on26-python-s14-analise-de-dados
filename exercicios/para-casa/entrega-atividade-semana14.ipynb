```python
#importando as bibliotecas 
import pandas as pd
```

# Coleta de dados:
Bases escolhidas para a atividade.
- clientes (olist_customers_dataset.csv)
- produtos (olist_products_dataset.csv)
- vendedores (olist_sellers_dataset.csv)

```python
df_clientes = pd.read_csv('olist_customers_dataset.csv')
df_clientes.info()
```

```python
df_produtos = pd.read_csv('olist_products_dataset.csv')
df_produtos.info()
```

```python
df_vendedores = pd.read_csv('olist_sellers_dataset.csv')
df_vendedores.info()
```

# Limpeza dos dados 
Removendo as colunas que não vamos usar.

```python
columns_drop_customer = ['customer_unique_id', 'customer_zip_code_prefix']
df_clientes = df_clientes.drop(columns_drop_customer, axis=1)
df_clientes
```

```python
columns_drop_products = ['product_description_lenght','product_photos_qty','product_weight_g','product_length_cm','product_height_cm','product_width_cm']

df_produtos = df_produtos.drop(columns=columns_drop_products, axis=1)
df_produtos
```

```python
columns_drop_sellers = ['seller_zip_code_prefix']
df_vendedores = df_vendedores.drop(columns_drop_sellers, axis=1)
df_vendedores
```

# Insights

```python
df_clientes_cidade = df_clientes['customer_city'].value_counts()
df_clientes_cidade.head(5)
```

```python
df_clientes_estado = df_clientes['customer_state'].value_counts()
df_clientes_estado.head(5)
```

```python

df_produtos_categoria = df_produtos['product_category_name'].value_counts()
produtos_top5 = df_produtos_categoria.head(5)

df_produtos_categoria
```

# Insights

- Qual a origem dos clientes da nossa basa de dados?

A maioria dos clientes é do estado de São Paulo e cidade de São Paulo; Seguido por Rio de Janeiro e cidade do Rio de Janeiro; E, por último, Belo Horizonte em Minas Gerais. 

- Quais são as categorias de produtos mais vendidos na nossa base de dados?

As top 5 categorias de produtos mais vendidos em nossa base de dados são: 1. cama, mesa e banho ; 2. esporte e lazer; 3. moveis e decoração; 4. beleza e saúde; 5. utilidades domésticas.

- Quais são as categorias de produtos menos vendidos na nossa base de dados?

As top 5 categorias de produtos menos vendidos em nossa base de dados são: 1. vestuário infanto-juvenil ; 2. casa e conforto ; 3. pc gamer ; 4. seguros e serviços ; 5. cds e dvds musicais.

- Quantos clientes foram registrados em SP para nossa base de dados?

Foram registrados 41.746 clientes em SP.

# Gráficos a partir de nossas bases de dados

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.barplot(x=produtos_top5.index, y=produtos_top5.values, hue=produtos_top5.index, legend=False)
plt.title('Comparação de Tipos de Categoria de Produtos')
plt.xlabel('Categoria')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)

plt.show()

```

```python
top_10_estados = df_clientes_estado.nlargest(10)

top_10_estados.plot(kind="bar", edgecolor='black')

plt.xlabel("Estado")
plt.ylabel("Quantidade")
plt.title("Top 10 estados com mais clientes")

plt.show()

```

```python
dados_SP = 41746
dados_RJ = 12852
dados_MG = 11635
dados_RS = 5466
dados_PR = 5045

labels = ['SP', 'RJ', 'MG', 'RS', 'PR']
sizes = [dados_SP, dados_RJ, dados_MG, dados_RS, dados_PR ]

colors = ['pink', 'gray', 'yellow', 'orange', 'purple']

explode = (0.1, 0, 0, 0, 0)

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Gráfico de Comparação entre os top 5 estados com mais clientes')
plt.show()
```
