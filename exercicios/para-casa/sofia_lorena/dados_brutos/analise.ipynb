# Coleta dos Dados

```python
import pandas as pd
```

- tabela de itens dos pedidos (olist_order_items_dataset)

```python
df_itens_pedido = pd.read_csv('olist_order_items_dataset.csv')
df_itens_pedido
```

- tabela de detalhes de pagamento dos pedidos (olist_order_payments_dataset)

```python
df_pagamento = pd.read_csv('olist_order_payments_dataset.csv')
df_pagamento
```

```python
df_olist_pedidos = pd.merge(df_itens_pedido, df_pagamento, on=['order_id'], how='inner')
df_olist_pedidos
```

- tabela de review dos usuários sobre os pedidos
(olist_order_reviews_dataset)

```python
df_olist_pedidos_uf = pd.merge(df_olist_pedidos, df_review, on=['order_id'], how='inner')
df_olist_pedidos_uf
```

# Objetivo da Análise Exploratória focada nos IDS:

perguntas com base nessas 3 tabelas:

- Quais são os produtos mais vendidos?
- Qual é o número médio de parcelas por tipo de pagamento?
- Qual é a relação entre a avaliação do produto e o tempo de entrega?
- O tempo limite de entrega das compras está sendo respeitado?

# Quais são os produtos mais vendidos?

```python
produtos_mais_vendidos = df_itens_pedido['product_id'].value_counts().head(10).reset_index()
produtos_mais_vendidos.columns = ['product_id', 'Quantidade de Vendas']
print(produtos_mais_vendidos)
```

```python
import matplotlib.pyplot as plt
```

```python
# criando um gráfico de barras focado no ID do produto
plt.figure(figsize=(10, 6))
plt.bar(produtos_mais_vendidos['product_id'], produtos_mais_vendidos['Quantidade de Vendas'], color='skyblue')
plt.xlabel('ID do Produto')
plt.ylabel('Quantidade de Vendas')
plt.title('Top 10 Produtos Mais Vendidos')
plt.xticks(rotation=45)
plt.show()
```

# Qual é o número médio de parcelas por tipo de pagamento?

```python
numero_medio_parcelas_por_tipo_pagamento = df_olist_pedidos.groupby('payment_type')['payment_installments'].mean().reset_index()
numero_medio_parcelas_por_tipo_pagamento['payment_installments'] = numero_medio_parcelas_por_tipo_pagamento['payment_installments'].astype(int)
print(numero_medio_parcelas_por_tipo_pagamento)
```

```python
# gráfico de barras
data = {'payment_type': ['boleto', 'cartão de crédito', 'cartão de débito', 'voucher'],
        'payment_installments': [1, 3, 1, 1]}

colors = ['#1f77b4', '#2ca02c', '#d62728', '#9467bd']

df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
bars = plt.bar(df['payment_type'], df['payment_installments'])

for bar, color in zip(bars, colors):
    bar.set_color(color)

plt.xlabel('Tipo de Pagamento')
plt.ylabel('Número Médio de Parcelas')
plt.title('Número Médio de Parcelas por Tipo de Pagamento')

# definindo os valores do eixo y como números inteiros
plt.yticks(range(int(max(df['payment_installments']) + 1)))

plt.show()
```

# Qual é a relação entre a avaliação do produto e o tempo de entrega?

```python
df_olist_pedidos_uf = pd.merge(df_review, df_itens_pedido, on='order_id', how='inner')

# convertendo as colunas de datas para o tipo datetime
df_olist_pedidos_uf['review_creation_date'] = pd.to_datetime(df_olist_pedidos_uf['review_creation_date'])
df_olist_pedidos_uf['shipping_limit_date'] = pd.to_datetime(df_olist_pedidos_uf['shipping_limit_date'])

# calculando o tempo de entrega em dias
df_olist_pedidos_uf['tempo_de_entrega'] = (df_olist_pedidos_uf['review_creation_date'] - df_olist_pedidos_uf['shipping_limit_date']).dt.days

# calculando a média da avaliação do produto por tempo de entrega
media_avaliacao_por_tempo = df_olist_pedidos_uf.groupby('tempo_de_entrega')['review_score'].mean().reset_index()

print(df_olist_pedidos_uf.head())
print(df_olist_pedidos_uf.dtypes)
```

# O tempo limite de entrega das compras está sendo respeitado?

```python
# calculando a diferença entre a data limite de envio e a data de criação da avaliação
df_olist_pedidos_uf['tempo_entrega_respeitado'] = (df_olist_pedidos_uf['review_creation_date'] - df_olist_pedidos_uf['shipping_limit_date']).dt.days <= 0

# verificando a porcentagem de pedidos em que o tempo de entrega foi respeitado
porcentagem_respeitado = df_olist_pedidos_uf['tempo_entrega_respeitado'].mean() * 100

print(f"A porcentagem de pedidos em que o tempo de entrega foi respeitado é: {porcentagem_respeitado:.2f}%")
```

```python
# gráfico de pizza
fig, ax = plt.subplots(figsize=(8, 8))
labels = ['Respeitado', 'Não Respeitado']
sizes = [porcentagem_respeitado, 100 - porcentagem_respeitado]
colors = ['green', 'red']
explode = (0.1, 0)  # destacando a fatia "Respeitado"

ax.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.2f%%', startangle=140)
ax.set_title('Respeito ao Tempo de Entrega: {:.2f}%'.format(porcentagem_respeitado))

plt.show()
```
