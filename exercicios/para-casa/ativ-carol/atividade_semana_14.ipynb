```python
import pandas as pd
```

```python
df_pagamentos = pd.read_csv('datasets_olist/olist_order_payments_dataset.csv')
df_pagamentos
```

### Quais colunas utiizarei

```python
df_orders = pd.read_csv('datasets_olist/olist_orders_dataset.csv')
df_orders
```

### Quais colunas irei usar
- **order_id**: Identificador do pedido. Será a nossa chave com a tabela de pagamentos;
- **order_status**: status  do pedido;


```python
df_produtos_reviews = pd.read_csv('datasets_olist/olist_order_reviews_dataset.csv')
df_produtos_reviews
```

### Quais colunas irei usar
- **review_id**: código de revisão;
- **order_id**: ID do pedido;
- **review_score**: pontuação da revisão;
- **review_comment_message**: revisar o título do comentário;

```python
columns_drop = ['payment_sequential']
df_pagamentos_drop = df_pagamentos.drop(columns_drop, axis=1)
df_pagamentos_drop
```

```python
columns_drop = ['review_comment_title',	'review_creation_date',	'review_answer_timestamp', 'review_id']
df_produtos_reviews_drop = df_produtos_reviews .drop(columns_drop, axis=1)
df_produtos_reviews_drop
```

```python

df_pagamento_produtos_reviews = pd.merge( df_pagamentos_drop,df_produtos_reviews_drop, on=['order_id'], how='inner')
df_pagamento_produtos_reviews
```

```python
columns_drop = ['customer_id','order_purchase_timestamp','order_approved_at','order_delivered_carrier_date','order_delivered_customer_date','order_estimated_delivery_date']
df_orders_drop =df_orders.drop(columns_drop, axis=1)
df_orders_drop
```

```python

df_pagamento_produtos_orders_e_reviews = pd.merge( df_pagamento_produtos_reviews,df_orders_drop,  on=['order_id'], how='inner')
df_pagamento_produtos_orders_e_reviews
```

```python
df_pagamento_produtos_orders_e_reviews.isnull().sum()
```

```python
df_pagamento_produtos_orders_e_reviews = df_pagamento_produtos_orders_e_reviews.dropna(subset=['review_comment_message'])
df_pagamento_produtos_orders_e_reviews.shape
```

```python
# Checar valores nulos no dataset
print("Valores nulos por coluna do dataframe:")
df_pagamento_produtos_orders_e_reviews.isnull().sum() # para vericar a quantidade de valores nulos 
```

```python
# importando os pacotes que iremos utilizar
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Qual tipo de pagamento ocorre mais?

# definir tamanho da figura
plt.figure(figsize=(8, 6))

# instanciar o tipo de plotagem, no caso, um countplot usando x como os tipos de pagamentos
sns.countplot(data = df_pagamento_produtos_orders_e_reviews, x = 'payment_type', color = 'chartreuse')

# definir texto do título e eixos x e y
plt.title('Contagem dos tipos de pagamentos do usuários')
plt.xlabel('Pagamentos do usuários')
plt.ylabel('Contagem')

plt.show()
```

```python
#Qual a quantidade de parcelas dessas comprass?

# definir tamanho da figura
plt.figure(figsize=(8, 6))

# instanciar o tipo de plotagem, no caso, um countplot usando x como os numeros de parcelas
sns.countplot(data = df_pagamento_produtos_orders_e_reviews, x = 'payment_installments', color = 'deeppink')

# definir texto do título e eixos x e y
plt.title('Contagem parcelas de pagamentos do usuários')
plt.xlabel('Parcelas Pagamentos do usuários')
plt.ylabel('Contagem')

plt.show()
```

```python
# Qual tipo de pagamento ocorre mais e qual a quantidade de parcelas dessas compras ?
# definir tamanho da figura
plt.figure(figsize=(10, 6))

# instanciar o tipo de plotagem, no caso, um countplot usando x como as quantidades de parcelas
# de plano com tonalidades das barras definidas pela tipode pagamento
custom_palette = "Set2"

sns.countplot(data = df_pagamento_produtos_orders_e_reviews,x = 'payment_installments',hue = 'payment_type', palette=custom_palette)

# definir texto do título e eixos x e y
plt.xticks(fontsize='large', fontweight='bold', rotation=45)


plt.title('Quentidade de parcelas por usuario  X Tipos de pagementos ')
plt.xlabel('Tipo de pagementos')
plt.ylabel('Contagem')

# plotar e definir texto da legenda

plt.legend(title = 'Tipo de pagementos')

plt.show()  
```

```python
#Qual o nivel de de satifação com o produto ocorre mais?

df_tipo_Review_Score = df_pagamento_produtos_orders_e_reviews.groupby('review_score').agg({'order_id': 'nunique'})
df_tipo_Review_Score.sort_values(by='order_id', ascending=False)
```

```python
#Contagem dos tipos comentários

df_tipo_Review_comment = df_pagamento_produtos_orders_e_reviews.groupby('review_comment_message').agg({'order_id': 'nunique'})
df_tipo_Review_comment.sort_values(by='order_id', ascending=False)
```

```python
#Qual tipo de comentarios(Muito bom', 'Bom', 'Recomendo) ocorre mais?
# definir tamanho da figura
plt.figure(figsize=(8, 6))
# instanciar o tipo de plotagem, no caso, um countplot usando x como os tipos de comentarios
sns.countplot(data=df_pagamento_produtos_orders_e_reviews, x='review_comment_message', color='mediumslateblue', order=['Muito bom', 'Bom', 'Recomendo'])

# definir texto do título e eixos x e y
plt.title('Contagem dos tipos de comentarios dos usuários')
plt.xlabel('Comentários dos usuários')
plt.ylabel('Contagem')

plt.show()
```

```python


# Selecionar os top 10 tipos de comentarios mais comuns
top_10_pagamentos = df_pagamento_produtos_orders_e_reviews['review_comment_message'].value_counts().nlargest(10).index

# Filtrar o DataFrame original para incluir apenas os top 10 tipos de comentarios mais comuns
df_top_10 = df_pagamento_produtos_orders_e_reviews[df_pagamento_produtos_orders_e_reviews['review_comment_message'].isin(top_10_pagamentos)]

# Definir tamanho da figura
plt.figure(figsize=(10, 6))


custom_palette = "Set3"

sns.countplot(data=df_top_10, x='review_score', hue='review_comment_message',  palette=custom_palette)

# Definir texto do título e eixos x e y
plt.title('Nivel de satisfação por usuário X Tipos de comentários (Top 10)')
plt.xlabel('Nivel de satisfação')
plt.ylabel('Contagem')

# Plotar e definir texto da legenda
plt.legend(title='Nivel de satisfação')

plt.show()
```

```python
# Qual  Status de pedido ocorre mais?
# definir tamanho da figura
plt.figure(figsize=(8, 6))
# instanciar o tipo de plotagem, no caso, um countplot usando x como os tipos de Status
sns.countplot(data=df_pagamento_produtos_orders_e_reviews, x='order_status', color='plum')
plt.xticks(fontsize='large', fontweight='bold', rotation=45)


# definir texto do título e eixos x e y
plt.title('Contagem do status dos pedidos',fontweight='bold')
plt.xlabel('Status pedido',fontweight='bold')
plt.ylabel('Contagem',fontweight='bold')

plt.show()
```

```python
#1. Qual Status de pedido de acordo com valor pago?
# Definir tamanho da figura
plt.figure(figsize=(15, 6))

# Filtrar os top 10 valores pagos
top10_valores_pagos = df_pagamento_produtos_orders_e_reviews.nlargest(10, 'payment_value')

# Instanciar o tipo de plotagem (gráfico de barras empilhadas)
sns.barplot(data=top10_valores_pagos, x='order_status', y='payment_value', hue='payment_value', palette='pastel', ci=None)

# Definir texto do título e eixos x e y
plt.title('Distribuição dos Valores Pagos por Status do Pedido', fontweight='bold')
plt.xlabel('Status do Pedido', fontweight='bold')
plt.ylabel('Valor Pago', fontweight='bold')

# Adicionar legenda
plt.legend(title='Valor Pago', bbox_to_anchor=(1, 1))

# Exibir o gráfico
plt.show()
```

```python
#1. Qual Status de pedido de acordo com valor pago?

# Definir tamanho da figura
plt.figure(figsize=(8, 5))

# Filtrar os top 10 valores pagos
top10_valores_pagos = df_pagamento_produtos_orders_e_reviews.nlargest(20, 'payment_value')

# Agrupar os dados por status do pedido e calcular a soma dos valores pagos em cada categoria
aggregated_data = top10_valores_pagos.groupby('order_status')['payment_value'].sum()

# Criar um gráfico de pizza
plt.pie(aggregated_data, labels=aggregated_data.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2'))

# Definir texto do título
plt.title('Distribuição dos Valores Pagos por Status do Pedido', fontweight='bold')

# Exibir o gráfico
plt.show()

```
