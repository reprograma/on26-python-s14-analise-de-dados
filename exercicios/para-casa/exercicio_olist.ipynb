#  Ánalise de dados da Basde de dados Olist

Base de dados disponibilizada pela Olist, uma empresa de e-commerce. Essa base de dados
é composta por 9 tabelas, porém para esse exercicio iremos utilizar apenas 5, que são:

- itens dos pedidos (olist_order_items_dataset)
- detalhes de pagamento dos pedidos (olist_order_payments_dataset)
- detalhes do consumidor que fez os pedidos (olist_customers_dataset)
- detalhes de geolocalização do consumidor (olist_geolocation_dataset)
- detalhes dos produtos (olist_products_dataset)
- detalhes dos vendedores (olist_sellers_dataset)

[_Fonte_](https://www.kaggle.com/datasets/olistbr/brazilian-ecommerce)

### Juntando as tabelas

```python
# Por padrão é realizado o merge dos dados com o inner join
# Jutando os df pedidos e itens de pedidos
df_olist_v0 = pd.merge(df_cons_pedidos, df_geoloc_consumidor, left_on='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix', how='inner')
df_olist_v0
```

```python
df_olist_v0.head()
```

```python
columns_drop = ['customer_zip_code_prefix']
df_olist_v0 = df_olist_v0.drop(columns_drop, axis=1)
df_olist_v0
```

```python
columns_drop = ['customer_unique_id','customer_city', 'customer_state']
df_olist_v0 = df_olist_v0.drop(columns_drop, axis=1)
df_olist_v0
```

```python
df_olist_v0 = pd.merge(df_olist_v0, df_vendedores, left_on='geolocation_zip_code_prefix', right_on='seller_zip_code_prefix', how='inner')
df_olist_v0
```

```python
columns_drop = ['seller_zip_code_prefix','seller_city', 'seller_state']
df_olist_v0 = df_olist_v0.drop(columns_drop, axis=1)
df_olist_v0
```

```python
df_olist_v = pd.merge(df_pag_pedidos, df_pedidos, on='order_id', how='inner')
df_olist_v
```

```python
df_olist_v = pd.merge(df_olist_v, df_olist_v, on='product_id', how='inner')
df_olist_v
```

```python
columns_drop = ['order_id_x','payment_sequential_x','payment_type_x','payment_installments_x','payment_value_x']
df_olist_v = df_olist_v.drop(columns_drop, axis=1)
df_olist_v
```

```python
columns_drop = ['order_item_id_x', 'payment_sequential_y','order_item_id_y', 'price_x', 'price_y','freight_value_y', 'seller_id_y', 'shipping_limit_date_y']
df_olist_v = df_olist_v.drop(columns_drop, axis=1)
df_olist_v
```

```python
#Suponha que você deseja tirar uma amostra de cada DataFrame
sample_fraction = 0.01

# Tire uma amostra do primeiro DataFrame (df_olist_v0)
df_olist_v0_amostra = df_olist_v0.sample(frac=sample_fraction, random_state=1)

# Tire uma amostra do segundo DataFrame (df_olist_v)
df_olist_v_amostra = df_olist_v.sample(frac=sample_fraction, random_state=1)
```

```python
df_olist_v0_amostra.head()
df_olist_v0_amostra.shape
```

```python
df_olist_v_amostra.head()
df_olist_v_amostra.shape
```

```python
# Realize a junção dos DataFrames de amostra
df_olist_amostra = pd.merge(df_olist_v0_amostra, df_olist_v_amostra, left_on='seller_id', right_on='seller_id_x', how='inner')
df_olist_amostra.head()
```

# Ánalises

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Ánalise
df_olist_v0
```

#### Quais são os 10 Estados com mais clientes?

```python
# Define o estilo de fundo
plt.style.use('dark_background')

# Cria uma figura
plt.figure(figsize=(10, 6))

# Ordena os estados geográficos com base na contagem de clientes e pega os 10 maiores
top_10_states = df_olist_v0['geolocation_state'].value_counts().head(10).index

# Gere um gráfico de contagem de clientes por estado geográfico
sns.countplot(data=df_olist_v0, x='geolocation_state', order=top_10_states)

# Adiciona os valores nas barras do gráfico
for p in plt.gca().patches:
    plt.gca().annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=10, color='white', xytext=(0, 5), textcoords='offset points')

# Define o título e rótulos dos eixos
plt.title('Top 10 Estados com Mais Clientes')
plt.xlabel('Estado Geográfico')
plt.ylabel('Contagem de Clientes')

# Rotaciona os rótulos do eixo x para melhor legibilidade
plt.xticks(rotation=45)

# Exibe o gráfico
plt.show()

```

##### Qual é Valor dos pagamentos por estado geográfico?

```python
# Agrupe os dados por estado geográfico e calcule a soma dos valores de pagamento
group_dados = df_olist_amostra.groupby('geolocation_state')['payment_value_y'].sum().reset_index()

# Crie o gráfico de barras empilhadas
plt.figure(figsize=(10, 6))
sns.barplot(data=group_dados, x='geolocation_state', y='payment_value_y', palette='viridis')
plt.title('Valor Total de Pagamento por Estado Geográfico')
plt.xlabel('Estado Geográfico')
plt.ylabel('Valor Total de Pagamento')
plt.show()
```

##### Qual é a relação entre o valor do frete (freight_value_x) e o tipo de pagamento (payment_type_y)?

```python
# Média do valor do frete por tipo de pagamento
avg_freight_by_payment = df_olist_amostra.groupby('payment_type_y')['freight_value_x'].mean()
print(avg_freight_by_payment)

# Gráfico de barras da média do valor do frete por tipo de pagamento
avg_freight_by_payment.plot(kind='bar', figsize=(10, 6))
plt.title('Média do Valor do Frete por Tipo de Pagamento')
plt.xlabel('Tipo de Pagamento')
plt.ylabel('Média do Valor do Frete')
plt.show()
```

##### Existe alguma sazonalidade nas datas de envio (shipping_limit_date_x) em relação ao valor do pagamento (payment_value_y)?

```python
# Converter as datas de envio para o formato de data
df_olist_amostra['shipping_limit_date_x'] = pd.to_datetime(df_olist_amostra['shipping_limit_date_x'])

# Gráfico de dispersão das datas de envio em relação ao valor do pagamento
plt.figure(figsize=(10, 6))
plt.scatter(df_olist_amostra['shipping_limit_date_x'], df_olist_amostra['payment_value_y'], alpha=0.5)
plt.title('Relação entre Data de Envio e Valor do Pagamento')
plt.xlabel('Data de Envio')
plt.ylabel('Valor do Pagamento')
plt.show()
```
