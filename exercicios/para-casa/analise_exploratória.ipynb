# Análise Exploratória (Olist)

Nessa Análise eu trago algumas datasets da e-commerce Olist. São elas:
- olis_order_items_dataset
- olist_products_dataset
- olist_sellers_dataset

através dos dados desses datasets, será feito o tratamento dos dados, a limpeza e resolução de alguns insights/perguntas
a partir da análise.
Também será visualizado em gráficos alguns dos insights que foram pensados.


```python
# Carregando a bibliotexa pandas,matplotlib

import pandas as pd
import matplotlib as plt

```

```python
# Lendo os Datasets 

df_order_i = pd.read_csv('olist_order_items_dataset.csv')
df_order_i
```

```python
df_products = pd.read_csv('olist_products_dataset.csv')
df_products
```

```python
df_sellers = pd.read_csv('olist_sellers_dataset.csv')
df_sellers
```

Após feita a leitura dos datasets escolhidos para a nálise, será feito a união dos três datasets num só
dataset. Isso acontece quando eu utilizo a  chave em comum entre elas e a função merge() da biblioteca pandas.



```python
# Utilizando a função  merge para união dos datasets

df_orderi_products = pd.merge(df_order_i, df_products, on=['product_id'], how='inner')
df_orderi_products
```

```python
df_final= pd.merge(df_orderi_products, df_sellers, on=['seller_id'], how='inner')
df_final
```

Depois de feita a integração entre as tabelas, será feita agora a limpeza e tratamento dos dados

```python
#Tratando e Limpandos dos dados

df_final.info()
```

```python
# Identificando os dados nulos 
df_final.isnull().sum()
```

Após a identificação dos dados nulos percebi que duas colunas que obtém alguns desses dados não fará diferença na minha análise. Por isso segui fazendo exclusão de algumas dessas colunas que não iria fazer diferença, para trabalhar com a menor quantidade de números nulos possível.

```python

#usando o método drop para excluir as colunas

columns_drop = ['product_name_lenght','product_description_lenght','product_photos_qty']
df_final = df_final.drop(columns_drop, axis=1)

# DataFrame final onde farei a análise exploratória
df_final
```

```python
# Trazendo o dataframe já com os dados nulos e as colunas que não irei precisar excluídas
df_final.isnull().sum()
```

Feita a limpeza dos dados e carregamento do dataset que irei analisar,
formulei as seguintes perguntas para serem trazidas através da análise

- Qual a categoria de produto mais solicitada em pedidos.
- Quantas vendas acima de R$ 500.0 foram feitas por um vendedor do estado do PR?
- Qual o vendedor fez mais vendas?
- Qual produto tem o preço mais barato?



```python
#Qual a categoria de produto mais solicitada em pedidos.
pergunta_1 = df_final['product_category_name'].value_counts()
pergunta_1

# resposta Cama/Mesa/Banho é a categoria mais solicitada, foram realizados 11115 pedidos com a categoria.
```

```python
#Quantas vendas acima de R$500 foram feitas por um vendedor no estado do PR?
pergunta_2 = df_final[(df_final['price'] > 500.0) & (df_final['seller_state'] == 'PR')].value_counts
pergunta_2



# Resposta: Foram encontrados 273 registros atendendo essa busca
```

```python
# Qual vendedor fez mais vendas?
pergunta3= df_final['seller_id'].value_counts()
pergunta3

#Resposta o vendedor que fez mais vendas 6560211a19b47992c3666cc44a7e94c0  com  2033 vendas realizadas.

    
```

```python
pergunta4 = df_final['price'].min()
pergunta4
# Resposta o produto com menor valor vendido custa 0.85 centavos.
```

Depois de respondidas as perguntas, trago a exibição de: 
- Gráfico Barra exibindo Número de vendas por Estado de cada Vendedor
- Gráfico exibindo top 3 de categorias que venderam mais itens
- Gráfico exibindo as categorias e o peso do produto mais pesado de cada uma

```python


#criando um objeto para guardar a contagem dos estados e trazendo em ordem alfabética
vendas_por_estado = df_final['seller_state'].value_counts().sort_index()

# gráfico Barra exibindo Número de Vendas por Estado de Cada Vendedor
vendas_por_estado.plot(kind='bar', color='tab:blue', title='Número de Vendas por Estado de Cada vendedor')

plt.xlabel('Estado do Vendedor')
plt.ylabel('Número de Vendas')
plt.show()
```

```python



# agrupando as categorias pela quantidade de itens mais vendidos dessa categoria
vendas_por_categoria = df_final.groupby(['product_category_name'])['order_item_id'].sum().reset_index()

# fiz um sort para ordenar o agrupamento acima trazendo como parâmetro a coluna de do Id dos itens
vendas_por_categoria = vendas_por_categoria.sort_values(by='order_item_id', ascending=False)

# aqui fiz um filtro para trazer apenas as 3 primeiras categorias em que foram vendidos a maior quantidade de itens
top_3_categorias = vendas_por_categoria.head(3)

#gráfico Barras exibindo o Top 3 de Categorias que mais tem itens vendidos
plt.figure(figsize=(10, 6))
plt.bar(top_3_categorias['product_category_name'], top_3_categorias['order_item_id'], color='tab:blue')
plt.title('Top 3 de categorias que venderam mais itens')
plt.xlabel('Categorias')
plt.ylabel('Quantidade de Itens Vendidos')
plt.show()
```

```python


#Produto mais pesado de cada categoria
df_peso_produto_por_categoria = df_final.groupby(['product_category_name'])['product_weight_g'].max()

# gráfico barra exibindo  as categorias e o peso do produto mais pesado de cada uma
plt.figure(figsize=(12, 6))
df_peso_produto_por_categoria.plot(kind='bar', color='tab:blue')
plt.title('Produtos mais pesado de cada categoria')
plt.xlabel('Categorias')
plt.ylabel('Pesp do produto em gramas')
plt.xticks(rotation=90)  
plt.show()
```
