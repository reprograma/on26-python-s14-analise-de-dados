Questão 2: quais são as principais categorias de venda?

```python
# Quero somente o Top 10, então vou filtrar os dados
# Primeiro, criando uma coluna com a frequência de cada categoria
df_produtos['Freq'] = df_produtos.groupby('product_category_name')['product_category_name'].transform('count')
```

```python
# Agora classificando os dados do menor para o maior
df_produtos.sort_values(by=["Freq"], ascending=False, inplace=True)
```

```python
# Removendo duplicatas, pois quero apenas as categorias individualizadas
df_produtos = df_produtos.drop_duplicates(subset='product_category_name', keep='first')
```

```python
# Agora quero apenas o top 10
top10_categorias = df_produtos[0:10]
top10_categorias
```

```python
# Por fim, a visualização
# definir tamanho da figura
plt.figure(figsize=(8, 6))

# instanciar o tipo de plotagem, no caso, um countplot usando x como os tipos de plano
top10_categorias.plot(kind="bar",x="product_category_name",y="Freq")

# definir texto do título e eixos x e y
plt.title('Top 10 categorias de produtos')
plt.xlabel('')
plt.ylabel('Vendas')
plt.legend().remove()

plt.show()
```

Questão 3: qual é a taxa de vendas de acordo com o período do ano?

```python
#garantindo que as colunas então com o type certo
df_FluxoCaixa['payment_value'] = df_FluxoCaixa['payment_value'].astype('float')
```

```python
df_FluxoCaixa['order_purchase_timestamp'] = df_FluxoCaixa['order_purchase_timestamp'].astype('datetime64[ns]')
```

```python
# #crio uma coluna mes_ano
df_FluxoCaixa['mes_ano'] = df_FluxoCaixa['order_purchase_timestamp'].map(lambda x: 100*x.year + x.month)
```

```python
# quero somente dados de um ano
amostra_vendas = df_FluxoCaixa[df_FluxoCaixa['order_purchase_timestamp'] >= "2018-01-01"]
```

```python
# estou agrupando para saber o valor total de vendas por mês
df_total_vendas = amostra_vendas.groupby(['mes_ano']).sum("payment_value")
```

```python
# visualizando dados agrupados
df_total_vendas
```

```python
# vou excluir dados que não vou utilizar da base amostral
df_total_vendas.drop("payment_sequential", axis = 1, inplace=True)
df_total_vendas.drop("payment_installments", axis = 1, inplace=True)
```

```python
# visualização das vendas ao longo do ano
df_total_vendas.plot()
plt.title("Total das vendas ao longo do ano 2018")
plt.legend().remove()
plt.xlabel("")
ticks, labels = plt.xticks()
plt.xticks(ticks, labels=['Dez 17','Fevereiro', 'Abril', "Junho", "Agosto", "Outubro", "Dezembro"]);
```
