# Análise visual

O local dos maiores e menores consumidores é o mesmo local dos vendedores?

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Consumidores por estado

plt.figure(figsize=(12, 6))

sns.countplot(data = df_juncao,
              x = "customer_state")

# definir texto do título e eixos x e y

plt.title("Consumidores por estado")
plt.xlabel("Estado")
plt.ylabel("Contagem")
```

```python
# Vendedores por estado

plt.figure(figsize=(12, 6))

sns.countplot(data = df_juncao,
              x = "seller_state")

# definir texto do título e eixos x e y

plt.title("Vendedores por estado")
plt.xlabel("Estado")
plt.ylabel("Contagem")
```

```python
#  Geolocalização por estado

plt.figure(figsize=(12, 6))

sns.countplot(data = df_juncao,
              x = "geolocation_state")

# definir texto do título e eixos x e y

plt.title("Geolocalização por estado")
plt.xlabel("Estado")
plt.ylabel("Contagem")
```

#
Pela  análise dos gráficos, os estados com mais e menores consumidores são também dos vendedores, com pouca diferença.
