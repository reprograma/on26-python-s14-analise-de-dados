```python
import pandas as pd
```

```python
#Base de dados: pedidos (olist_orders_dataset)
df_pedidos = pd.read_csv('dados brutos/olist_orders_dataset.csv')
df_pedidos
```

```python
#Base de dados: itens dos pedidos (olist_order_items_dataset)
df_itens_pedido = pd.read_csv('dados brutos/olist_order_items_dataset.csv')
df_itens_pedido
```

```python
#Base de dados: review dos usuários sobre os pedidos (olist_order_reviews_dataset)
df_avaliacao_pedido = pd.read_csv('dados brutos/olist_order_reviews_dataset.csv')
df_avaliacao_pedido
```

```python
#Base de dados:detalhes dos vendedores (olist_sellers_dataset)
df_vendedores = pd.read_csv('dados brutos/olist_sellers_dataset.csv')
df_vendedores
```

 - Filtrando somente as colunas que vou utilizar:

```python
columns_drop = ['order_approved_at', 'order_delivered_carrier_date', 'order_estimated_delivery_date']
df_pedidos = df_pedidos.drop(columns_drop, axis=1)
df_pedidos
```

```python
columns_drop = ['order_id', 'order_item_id', 'seller_id']
df_itens_pedido = df_itens_pedido.drop(columns_drop, axis=1)
df_itens_pedido
```

```python
columns_drop = ['order_id', 'review_comment_title']
df_avaliacao_pedido = df_avaliacao_pedido.drop(columns_drop, axis=1)
df_avaliacao_pedido
```

```python
columns_drop = ['seller_zip_code_prefix']
df_vendedores = df_vendedores.drop(columns_drop, axis=1)
df_vendedores
```

- Filtrando informações específicas. Por exemplo: a nota atribuída na avaliação do produto, nesse caso quero visualizar somente a nota 5. 

```python
df_avaliacao_pedido = df_avaliacao_pedido.loc[df_avaliacao_pedido['review_score'] == 5]
df_avaliacao_pedido
```

```python
df_vendedores= df_vendedores.loc[df_vendedores['seller_state'] == 'AM']
df_vendedores
```
