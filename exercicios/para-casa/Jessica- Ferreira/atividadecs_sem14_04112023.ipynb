- Qual o tipo de pagamento mais usado nas compras?
A categoria mais utilizada foi o cartão de crédito, utilizado 84.732 vezes de 114.859 compras analisadas nessa base.

```python
tipo_pagamento = df_olist_exerc["payment_type"].value_counts()
tipo_pagamento.plot.bar(edgecolor='black');

plt.xlabel("")
plt.ylabel("")
plt.title("Tipo de pagamento usado nas compras")

tipo_pagamento
```

- Qual a porcentagem por tipo de pagamento?
Aproximadamente 1% do total das compras são realizadas pelocartão de crédito.

```python
tipo_pagamento_por100 = df_olist_exerc["payment_type"].value_counts(normalize=True)
tipo_pagamento_por100
```

- Que estado mais utiliza o cartão de crédito como forma de pagamento?
São Paulo é o estado que mais utiliza o cartão de crédito, seguido do Rio de Janeiro e Minas Gerais.

```python
plt.figure(figsize=(12, 6))
sns.countplot(data = df_olist_exerc,
              x = 'customer_state',
              hue = 'payment_type')

plt.show()
```

- Que estado mais recebeu a pontuação máxima em suas avaliações?
São Paulo foi o estado com mais pontuações, mas podemos ver uma diferença de mais 18.000 em relação a quantidade de compras que foi de 48.517. 

```python
plt.figure(figsize=(12, 6))
sns.countplot(data = df_olist_exerc,
              x = "customer_state",
              hue = "review_score")

plt.title('Estados que mais avaliam suas compras')
plt.xlabel('Estados')
plt.ylabel('')
plt.legend(title = 'Estrelas')

plt.show()
```
