```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df_pagamentos = pd.read_csv('olist_order_payments_dataset.csv')
df_pagamentos
```

```python
df_pagamentos.info()
```

```python
columns_drop = ['payment_sequential', 'payment_installments']
df_pagamentos = df_pagamentos.drop(columns_drop, axis=1)
df_pagamentos
```

```python
df_pedidos = pd.read_csv('olist_orders_dataset.csv')
df_pedidos
```

```python
df_pedidos.info()
```

```python
columns_drop = ['order_estimated_delivery_date', 'order_delivered_customer_date', 'order_delivered_carrier_date','order_purchase_timestamp']
df_pedidos = df_pedidos.drop(columns_drop, axis=1)
df_pedidos
```

```python
df_itens_pedido = pd.read_csv('olist_order_items_dataset.csv')
df_itens_pedido
```

```python
df_itens_pedido.info()
```

```python
df_consumidor = pd.read_csv('olist_customers_dataset.csv')
df_consumidor
```

```python
df_consumidor = df_consumidor.drop('customer_unique_id', axis=1)
```

```python
df_localizacao = pd.read_csv('olist_geolocation_dataset.csv')
df_localizacao
```

```python
df_pagamentos.info()
```

```python
df_juncaoPedidos = pd.merge(df_pagamentos, df_pedidos, on= 'order_id')
df_juncaoPedidos
```

```python
# Qual o payment_type mais prevalente?
contagem_pagamentos = df_pagamentos['payment_type'].value_counts()
contagem_pagamentos
```

```python
contagem_pagamentos.plot(kind='bar')
plt.title('Contagem de Pagamentos por Tipo')
plt.xlabel('Tipo de Pagamento')
plt.ylabel('Contagem')
plt.show()
```

```python
# Qual a maior compra payment_value?
maior_compra = df_pagamentos['payment_value'].max()
maior_compra
```

```python
maior_compra = df_pagamentos.loc[df_pagamentos['payment_value'].idxmax()]

plt.bar(maior_compra['order_id'], maior_compra['payment_value'])
plt.title('Maior Compra')
plt.xlabel('ID da Compra')
plt.ylabel('Valor da Compra')
plt.show()
```

```python
# Qual Estado que mais compra?
consumidor_estado = df_localizacao['geolocation_state'].value_counts()
consumidor_estado
```

```python
consumidor_estado.plot(kind='bar')
plt.title('Contagem de Consumidor por Estado')
plt.xlabel('Estados')
plt.ylabel('Contagem')
plt.show()
```
