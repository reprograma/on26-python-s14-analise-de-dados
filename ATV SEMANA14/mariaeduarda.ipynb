```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/csvs/Cópia de Cópia de olist_customers_dataset.csv')
```

```python
df.head()
```

```python
#Qual a cidade com mais clientes?
customer_city = df['customer_city'].value_counts().nlargest(5)
plt.bar(customer_city.index, customer_city.values)
plt.xlabel('Cidade')
plt.ylabel('Número de clientes')
plt.title('Top 5 Cidades com mais clientes')
plt.xticks(rotation=45)
plt.show()
```

```python
df2 = pd.read_csv('/content/drive/MyDrive/csvs/Cópia de olist_products_dataset.csv')
df2.head()
```

```python
#Qual a categoria de produtos mais vendida?
product_category_name = df2['product_category_name'].value_counts().nlargest(5)
plt.bar(product_category_name.index, product_category_name.values)
plt.xlabel('Categoria')
plt.ylabel('Quantidade')
plt.title('Top 5 categorias mais vendidas')
plt.xticks(rotation=45)
plt.show()
```
