```python
df_vendas = pd.read_csv('olist_order_items_dataset.csv')
df_vendas
```

```python
columns_drop1 = ['order_item_id','seller_id']
df_vendas = df_vendas.drop(columns_drop1, axis=1)
df_vendas
```

```python
df_vendas.info()
```

```python
#Tranformanda a coluna shipping_limit_date em data
df_vendas['shipping_limit_date'] = pd.to_datetime(df_vendas['shipping_limit_date'])
df_vendas.info()
```

***<div style="background-color:white; padding:10px;text-align:center"><span style="color:purple; font-size:1em">Analisando a tabela de dados (Avaliações)***<img src="https://pa1.aminoapps.com/7094/61a0f7dd0a3db2608cfdc46140deff3f544307f9r1-256-256_00.gif" alt="Super Mario Bros" width="50"/></span></div>

```python
df_avaliacoes = pd.read_csv('olist_order_reviews_dataset.csv')
df_avaliacoes
```

```python
df_avaliacoes.info()
```

```python
columns_drop2 = ['review_id','review_comment_title']
df_avaliacoes = df_avaliacoes.drop(columns_drop2, axis=1)
df_avaliacoes
```

***<div style="background-color:white; padding:10px;text-align:center"><span style="color:purple; font-size:1em">Convertendo colunas do df_avaliacoes que recebem datas***<img src="https://pa1.aminoapps.com/7094/61a0f7dd0a3db2608cfdc46140deff3f544307f9r1-256-256_00.gif" alt="Super Mario Bros" width="50"/></span></div>

```python

df_avaliacoes['review_creation_date'] = pd.to_datetime(df_avaliacoes['review_creation_date'])
df_avaliacoes['review_creation_date'].dt.day

df_avaliacoes['review_answer_timestamp'] = pd.to_datetime(df_avaliacoes['review_answer_timestamp'])
df_avaliacoes['review_answer_timestamp'].dt.day
df_avaliacoes.info()
```

***<div style="background-color:white; padding:10px;text-align:center"><span style="color:purple; font-size:1em">Avaliando a tabela (Ordens dos Pedidos)***<img src="https://pa1.aminoapps.com/7094/61a0f7dd0a3db2608cfdc46140deff3f544307f9r1-256-256_00.gif" alt="Super Mario Bros" width="50"/></span></div>

```python
df_pedidos = pd.read_csv('olist_orders_dataset.csv')
df_pedidos
```

```python
df_pedidos.info()
```

***<div style="background-color:white; padding:10px;text-align:center"><span style="color:purple; font-size:1em">Transformando em datas colunas do df_pedidos***<img src="https://pa1.aminoapps.com/7094/61a0f7dd0a3db2608cfdc46140deff3f544307f9r1-256-256_00.gif" alt="Super Mario Bros" width="50"/></span></div>

```python
df_pedidos['order_purchase_timestamp'] = pd.to_datetime(df_pedidos['order_purchase_timestamp'])
df_pedidos['order_purchase_timestamp'].dt.day
df_pedidos['order_approved_at'] = pd.to_datetime(df_pedidos['order_approved_at'])
df_pedidos['order_approved_at'].dt.day
df_pedidos['order_delivered_carrier_date'] = pd.to_datetime(df_pedidos['order_delivered_carrier_date'])
df_pedidos['order_delivered_carrier_date'].dt.day
df_pedidos['order_delivered_customer_date'] = pd.to_datetime(df_pedidos['order_delivered_customer_date'])
df_pedidos['order_delivered_customer_date'].dt.day
df_pedidos['order_estimated_delivery_date'] = pd.to_datetime(df_pedidos['order_estimated_delivery_date'])
df_pedidos['order_estimated_delivery_date'].dt.day
df_pedidos.info()

```

***<div style="background-color:white; padding:10px;text-align:center"><span style="color:purple; font-size:2em">Agregando os Dataset***<img src="https://pa1.aminoapps.com/7094/61a0f7dd0a3db2608cfdc46140deff3f544307f9r1-256-256_00.gif" alt="Super Mario Bros" width="50"/></span></div>

```python
df_olist1 = pd.merge(df_pagamentos, df_vendas, on='order_id')
df_olist1
```

```python
df_olist2 = pd.merge(df_olist1, df_avaliacoes, on='order_id')
df_olist2
```

```python
df_olist3 = pd.merge(df_olist2, df_pedidos, on='order_id')
df_olist3
```

```python
df_olist4 = pd.merge(df_olist3, df_cliente, on='customer_id')
df_olist4
```

```python
df_olist4['reference_month'] = df_olist4['order_purchase_timestamp'].dt.strftime('%Y-%B')
df_olist4
```

***<div style="background-color:white; padding:10px;text-align:center"><span style="color:purple; font-size:2em">Limpando os dados***<img src="https://pa1.aminoapps.com/7094/61a0f7dd0a3db2608cfdc46140deff3f544307f9r1-256-256_00.gif" alt="Super Mario Bros" width="50"/></span></div>

```python

columns_drop = ['order_id','product_id','customer_id']
df_olist4 = df_olist4.drop(columns_drop, axis=1)
df_olist4
```

```python
volume_de_Venda_por_estados = df_olist4['customer_state'].value_counts()

volume_de_Venda_por_estados
```

```python
plt.figure(figsize=(10, 6))
volume_de_Venda_por_estados.plot(kind='bar', color='blue')


plt.title('VOLUME DE VENDAS POR ESTADO')
plt.xlabel('ESTADOS')
plt.ylabel('VOLUME DE VENDAS')

plt.show()
```

***<div style="background-color:purple; padding:10px;text-align:center"><span style="color:yellow; font-size:2em">O Estado com maior volume de Vendas foi São Paulo com 49.527 vendas e o menor volume foi identificado em Roraima com apenas 52 vendas***<img src="https://pa1.aminoapps.com/7094/61a0f7dd0a3db2608cfdc46140deff3f544307f9r1-256-256_00.gif" alt="Super Mario Bros" width="50"/></span></div>

***<div style="background-color:blue; padding:10px;text-align:center"><span style="color:yellow; font-size:2em">Forma de pagamento mais utilizada***<img src="https://pa1.aminoapps.com/7094/61a0f7dd0a3db2608cfdc46140deff3f544307f9r1-256-256_00.gif" alt="Super Mario Bros" width="50"/></span></div>

```python
forma_de_pagamento = df_olist4['payment_type'].value_counts()
forma_de_pagamento
```

```python

plt.figure(figsize=(6, 6))  
forma_de_pagamento.plot(kind='pie', autopct='%1.1f%%', startangle=50) 
plt.title('FORMAS DE PAGAMENTOS MAIS UTILIZADAS')  # Título do gráfico
plt.ylabel('')
plt.show()  # Exibe o gráfico
```

***<div style="background-color:purple; padding:10px;text-align:center"><span style="color:yellow; font-size:2em">O meio de pagamento mais utilizado é o cartão de crédito e o menos é o cartão de débito***<img src="https://pa1.aminoapps.com/7094/61a0f7dd0a3db2608cfdc46140deff3f544307f9r1-256-256_00.gif" alt="Super Mario Bros" width="50"/></span></div>

***<div style="background-color:blue; padding:10px;text-align:center"><span style="color:yellow; font-size:2em">Avaliação dos clientes***<img src="https://pa1.aminoapps.com/7094/61a0f7dd0a3db2608cfdc46140deff3f544307f9r1-256-256_00.gif" alt="Super Mario Bros" width="50"/></span></div>

```python
score_av = df_olist4['review_score'].value_counts().sort_index(ascending=True)
score_av
```

```python

plt.figure(figsize=(8, 6))  
score_av.plot(kind='bar', color='yellow')  
plt.xlabel('Classificação') 
plt.ylabel('Quantidade de Avaliações') 
plt.title('AVALIAÇÕES')  
plt.xticks(rotation=0) 
plt.tight_layout()  
plt.show()  
```

***<div style="background-color:purple; padding:10px;text-align:center"><span style="color:yellow; font-size:2em">Podemos classificar a empresa com um bom atendimento, já que a maioria dos clientes deram nota máxima (5) para o serviço prestado***<img src="https://pa1.aminoapps.com/7094/61a0f7dd0a3db2608cfdc46140deff3f544307f9r1-256-256_00.gif" alt="Super Mario Bros" width="50"/></span></div>
